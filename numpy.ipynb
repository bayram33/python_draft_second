###  6=> Iterating over Arrays

```python
import numpy as np
```

```python
a=np.arange(11)**2
a
```

```python
for i in a:
    print(i**(1/2))
```

```python
student = np.array(
    [
        ["alice", "beth", "cathy", "dorothy"],
        [12, 32, 44, 11],
        [98, 52, 56, 71],
    ]
)
```

```python
for x in student:
    print(x)
```

```python
for x in student.flatten():
    print(x)
```

```python
for x in student.flatten(order="F"):
    print(x)
```

```python
# flatten actually returns a result array, nditer just allows you to iterate over elements
x=np.arange(12).reshape(3,4)
x
```

```python
print(x)
print("***")
for element in np.nditer(x):
    print(element)
```

```python
print(x)
print("***")
for element in np.nditer(x , order="F"):
    print(element)
```
