### Aggregate
* aggregate fonksiyonu, gruplar halindeki veriler üzerinde toplu işlemler yapmanızı sağlar. Bu işlemler toplam, ortalama, maksimum, minimum vb. gibi istatistiksel işlemleri içerebilir. 

```python
import pandas as pd

df = pd.DataFrame(
    {
        "Class": ["A", "B", "C", "A", "B", "C"],
        "Variable1": [12, 32, 12, 12, 43, 65],
        "Variable2": [33, 76, 2, 46, 9, 3],
    }
)
df
```

```python
df.groupby("Class").agg(["min","max","sum","median","mean"])
```

```python
df.groupby("Class").aggregate(["min","max","sum","median"])
```

```python
df.aggregate(
    {
        "Variable1":["sum","min"],
        "Variable2":["max","min","sum"]
    }
)
```

```python
df.groupby("Class").aggregate(
    {
        "Variable1":["sum","min"],
        "Variable2":["max","min","sum"]
    }
)
```

### Filter
* filter fonksiyonu, belirli bir koşulu sağlayan grupları seçmenizi sağlar. Örneğin, bir grup içindeki verilerin sayısı belirli bir eşik değerin üzerinde mi altında mı kontrol edebilirsiniz:

```python
import pandas as pd


df = pd.DataFrame(
    {
        "class": ["A", "B", "C", "A", "B", "C"],
        "data1": [7, 4, 1, 9, 5, 6],
        "data2": [1, 2, 3, 4, 5, 6],
        "data3": [2, 3, 3, 4, 2, 9],
    },
)
df
```

```python
df.groupby("class").std()
```

```python
df.groupby("class").filter(lambda x: x["data1"].std() < 1)
```

### Transform
* transform fonksiyonu, her grup için bağımsız bir dönüşüm işlemi yapmanızı sağlar. Örneğin, her grup içindeki değerleri kendi grup ortalamasıyla normalize etmek için kullanılabilir:

```python
import pandas as pd


df = pd.DataFrame(
    {
        "class": ["A", "B", "C", "A", "B", "C"],
        "data1": [7, 4, 1, 9, 5, 6],
        "data2": [1, 2, 3, 4, 5, 6],
        "data3": [2, 3, 3, 4, 2, 9],
    },
)
df
```

```python
df.groupby("class").mean()
```

```python
df.groupby("class").transform(lambda x: x - x.mean())
```

### Apply
* apply fonksiyonu, DataFrame'in her satırına veya sütununa belirli bir işlem uygulamanızı sağlar. Bu işlem, özel bir işlev veya lambda işlevi olabilir. Örneğin, bir sütundaki her değeri karekök almak için kullanılabilir:

```python
import pandas as pd


df = pd.DataFrame(
    {
        "class": ["A", "B", "C", "A", "B", "C"],
        "data1": [7, 4, 1, 9, 5, 6],
        "data2": [1, 2, 3, 4, 5, 6],
        "data3": [2, 3, 3, 4, 2, 9],
    },
)
df
```

```python
df.groupby("class").apply(lambda x: x["data1"] / x["data2"])
```

```python
def by_data(x):
    x["data1"] /= x["data2"]
    return x

df.groupby("class").apply(by_data)
```

# Pivot Table

```python
import pandas as pd 
import seaborn as sns 
titanic= sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.groupby("sex")["survived"].mean()
```

```python
titanic.groupby("sex")[["survived"]].mean()
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

```python
# with pivot table
titanic.pivot_table(values="survived", index="sex", columns="class")
```
